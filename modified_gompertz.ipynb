```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,200,100)
P_m = 52.47
R_m = 0.52
L = 6.31
#P = P_m*np.exp(-np.exp(R_m*np.e/P_m)*(L-t)+1)
# P = P_m * np.exp(-np.exp((R_m * np.e / P_m) * (L - t)) + 1)
P = P_m * np.exp(-np.exp((R_m * np.e / P_m) * (L - t) + 1))

plt.plot(t, P)
plt.title("Ethanol Produced Over Time")
plt.xlabel("Fermentation Time, $t$ (h)")
plt.ylabel("Ethanol Concentration, $P$ (g/L)")
plt.grid()
plt.show()
```
