<h1> Euler's Method </h1>

Euler's Method is used to generate approximate values of the solution to an initial value problem (IVP) of the form 

$\displaystyle{{dy \over dx}=G(x,y)}$,  where $y(x_0)=y_0$. 

<b> <u> Algorithm: </u>  </b>

$x_0$ and $y_0$ are given by the initial condition. To compute successive approximations, use:

<p style="color:Blue;">$x_{n}=x_{n-1} + h$ </p>

<p style="color:Blue;">$y_{n}=y_{n-1} + G(x_{n-1},y_{n-1})\cdot h$</p>

where $n=1,2,3,4$ and so on, and $h$ is the step size (sometimes called $\Delta x$).


```python
## basic code for Euler's method

# IVP: dy/dx = x + y, y(0)=1

n = 10 # number of iterations/steps
h = 1  # step size

x = [0] # array for x, with initial condition x_0 entered
y = [1] # array for y, with initial condition y_0 entered

for i in range(1, n+1):
    xi = x[i-1] + h
    yi = y[i-1] + (x[i-1]+y[i-1])*h # don't forget to modify this line!
    x.append(xi)
    y.append(yi)
    print("(",xi,",",yi,")") # you may wish to comment out this line as n becomes large

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Approximate Solution Using Euler's Method")
plt.grid()
```

**Logistic Model**

<u>Recall</u>: The logistic model describes a population that grows exponentially for small values of $P$ but as $P$ increases, the growth rate slows down and the population approaches the carrying capacity. If the population starts above its carrying capacity, it will decrease towards the carrying capacity.

<p style="color:Blue;"><b> <u> Logistic Differential Equation </b> </u></p>

$\displaystyle{ {dP \over dt} = kP\left(1-{P \over L}\right)}$

where $k$ is a positive constant and $L$ is the carrying capacity.

**Predator-Prey Model**

The logistic differential equation and Gompertz equation model the growth of a single species that lives alone in an environment. Now, we will consider two species interacting in the same habitat. One species, called the <b> prey</b>, has an ample food supply and the second species, called the <b> predator</b>, feeds on the prey.

For example, consider a population of rabbits (prey) and wolves (predators) in an isolated
forest. 

Let $\displaystyle{R(t)}$ = # of rabbits at time $t$ and $\displaystyle{W(t)}$ = # of wolves at time $t$.

<u>Assumption 1:</u>
Without predators, prey will grow exponentially:

$\displaystyle{ {dR(t) \over dt} = kR}$, $k>0$

Without prey, predators will die out exponentially:

$\displaystyle{ {dW(t) \over dt} = -rW}$, $r>0$


<u>Assumption 2:</u>
There will be more encounters between the two species if the population of either increases:

number of encounters $\propto RW$

<u>Assumption 3:</u>
Encounters are bad for prey:

$\displaystyle{ {dR(t) \over dt} = kR-aRW}$, $a>0$

Encounters are good for predators:

$\displaystyle{ {dW(t) \over dt} = -rW+bRW}$, $b>0$

Thus, we have developed the following system of differential equations.

<p style="color:Blue;"><b> <u> Predator-Prey Equations or Lotka-Volterra Equations</b> </u></p>

$\displaystyle{ {dR(t) \over dt} = kR-aRW}$

$\displaystyle{ {dW(t) \over dt} = -rW+bRW}$

where $k, r, a, b$ are positive constants.


```python
## predator-prey model; p. 650 in your text

n = 12
h = 1 
k = 0.08
r = 0.02
a = 0.001
b = 0.00002
print(k*1000 - a*1000*80)
print(-r*80 + b*1000*80)

t = [0] # array for t with initial condition t_0 entered
R = [500] # array for R with initial condition R_0 entered
W = [100] # array for W with initial condition W_0 entered

for i in range(1, n+1):
    ti = t[i-1] + h
    Ri = R[i-1] + (k*R[i-1]-a*R[i-1]*W[i-1])*h 
    Wi = W[i-1] + (-r*W[i-1]+b*R[i-1]*W[i-1])*h
    t.append(ti)
    R.append(Ri)
    W.append(Wi)
    #print("(", ti, "," , round(Ri), ",", round(Wi), ")")

# plot solution
plt.plot(t,R, label = 'rabbits')
plt.plot(t,W, label = 'wolves')
plt.title("Approximate Solution to Predator-Prey Model - Euler's Method")
plt.xlabel("time")
plt.ylabel("population")
plt.legend()
plt.grid()
plt.show()

# semi-log plot 
plt.plot(t,np.log(R), label = 'rabbits')
plt.plot(t,np.log(W), label = 'wolves')
plt.title("Approximate Solution to Predator-Prey Model - Euler's Method")
plt.xlabel("time")
plt.ylabel("ln(population)")
plt.legend()
plt.grid()
plt.show()
```

<h1><font color='navy'> Riemann Sums </font></h1>

The definite integral of a function $f(x)$ from $x=a$ to $x=b$ is defined as the limit of a Riemann sum:

$\displaystyle{\, \int_{a}^b f(x) dx\,\,}= \lim_{n\to\infty} \sum_{i=1}^{n} f(x^*_i)\Delta x$

where $\displaystyle{\Delta x = {{b-a} \over n}}$ and $x^*_i$ is any sample point in the subinterval $[x_{i-1},x_i]$.

**Note 1**: We can approximate definite integrals by using a <i> finite </i> Riemann sum. In general, the larger the value of $n$, the closer the Riemann sum is to the actual value of the definite integral. Also, as $n$ becomes larger and larger, the choice of sample point $x^*_i$ has less of an impact on the overall sum. <i> **Why?** </i>


**Note 2**: When a function $f(x)$ is non-negative on an interval $[a,b]$, we can interpret 
$\displaystyle{\, \int_{a}^b f(x) dx\,\,}$ as the area of the region under the curve $f(x)$ over the interval $[a,b]$. We can approximate the area of this region using a finite Riemann sum.


**Computing Sums in Python**

```python
## Riemann sums - approximating the area under y=x^2 on [0,2]; compare to calculations above

# define parameters
n = 4 # number of rectangles
a = 0 # left endpoint
b = 2 # right endpoint

# define variables  
x = np.linspace(a,b,n+1) # creates an array of n+1 equally spaced x-values between a and b
y = x**2 # define y=f(x) here

# compute the width of each rectangle
w = (b-a)/n

# compute the areas of all possible rectangles
a = y*w 
print("The a-array stores the areas of", len(a), "approximating rectangles.")

# compute the left-sum (a[0]+a[1]+...+a[n-1])
ls = sum(a[0:n])
print("The left sum is:",ls)

# compute right-sum (a[1]+a[2]+...+a[n])
rs = sum(a[1:n+1])
print("The right sum is:",rs)
```

**Computing Riemann sums - version 2**

Study the code in the cell below which computes $L_4$ and $R_4$ for $y=x^2$ on $[0,2]$, using arrays and loops.

Add to the code to find $M_4$. How does it relate to $L_4$ and $R_4$?

```python
## Riemann sums, code using arrays and loops

# define parameters
n = 4
a = 0
b = 2
delta_x = (b-a)/n

# define variables
x = np.linspace(a,b,n+1) 
y_left = []  # define array for y-values at left endpoints of each subinterval
y_right = [] # define array for y-values at right endpoints of each subinterval
y_midpoint = [] # define array for y-values at midpoints of each subinterval

# left sum
for i in range(n):
    xl = x[i]
    yl = xl**2
    y_left.append(yl)
ls = sum(y_left)*delta_x
print("The left sum is:", ls)

# right sum
for i in range(n):
    xr = x[i+1]
    yr = xr**2
    y_right.append(yr)
rs = sum(y_right)*delta_x
print("The right sum is:", rs)

# midpoint sum 
for i in range(n):
    xm = (x[i]+x[i+1])/2
    ym = xm**2
    y_midpoint.append(ym)
ms = sum(y_midpoint)*delta_x
print("The midpoint sum is:", ms)
```
